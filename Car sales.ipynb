```python
#1 What is the average selling price of cars for each dealer, and how does it compare across different dealers.

avg_selling_price = df.groupby('Dealer_Name')['Price ($)'].mean().reset_index()

    # Rename columns for clarity
avg_selling_price.columns = ['Dealer Name', 'Average Selling Price']

    # Sort by average selling price for better comparison
avg_selling_price = avg_selling_price.sort_values(by='Average Selling Price', ascending=False)

    # Display the results
print(avg_selling_price)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(avg_selling_price['Dealer Name'], avg_selling_price['Average Selling Price'], color='skyblue', edgecolor='black')
plt.title('Average Selling Price of Cars by Dealer')
plt.xlabel('Dealer Name')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
price_variation = df.groupby('Company')['Price ($)'].std().reset_index()

    # Rename columns for clarity
price_variation.columns = ['Company', 'Price Standard Deviation']

    # Sort by standard deviation to find the brand with the highest variation
price_variation = price_variation.sort_values(by='Price Standard Deviation', ascending=False)

    # Display the brand with the highest variation
highest_variation_brand = price_variation.iloc[0]
print(f"Company with the highest price variation: {highest_variation_brand['Company']}")
print(f"Price Standard Deviation: {highest_variation_brand['Price Standard Deviation']:.2f}")

    # Display the full table of price variations
print("\nPrice Variation by Company:")
print(price_variation)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(price_variation['Company'], price_variation['Price Standard Deviation'], color='skyblue', edgecolor='black')
plt.title('Price Variation by Car Brand')
plt.xlabel('Car Brand')
plt.ylabel('Price Standard Deviation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges
#compare.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Transmission', y='Price ($)', data=df, palette='pastel')
plt.title('Distribution of Car Prices by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Selling Price')
plt.tight_layout()
plt.show()

    # Calculate IQR for each transmission type
iqr_data = df.groupby('Transmission')['Price ($)'].describe(percentiles=[0.25, 0.75])
iqr_data['IQR'] = iqr_data['75%'] - iqr_data['25%']

    # Display IQR comparison
print("\nInterquartile Range (IQR) for Each Transmission Type:")
print(iqr_data[['IQR']])
```

```python
#4. What is the distribution of car prices across different regions.
data = df.dropna(subset=['Dealer_Region', 'Price ($)'])

# Plot a box plot to visualize the distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='Dealer_Region', y='Price ($)', data=data, palette="coolwarm")
plt.title('Distribution of Car Prices Across Regions')
plt.xlabel('Region')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#5. What is the distribution of cars based on body styles.

body_style_counts = data['Body Style'].value_counts()

# Display the counts
print(body_style_counts)

# Plot a bar chart for distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=body_style_counts.index, y=body_style_counts.values, palette="viridis")
plt.title('Distribution of Cars Based on Body Styles')
plt.xlabel('Body Style')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#6. How does the average selling price of cars vary by customer gender and annual income.

income_bins = [0, 25000, 50000, 75000, 100000, np.inf]
income_labels = ['<25k', '25-50k', '50-75k', '75-100k', '>100k']
data['income_group'] = pd.cut(data['Annual Income'], bins=income_bins, labels=income_labels)

# Group by gender and income group, and calculate the average selling price
grouped_data = data.groupby(['Gender', 'income_group'])['Price ($)'].mean().reset_index()

# Pivot the data for easier visualization
pivot_data = grouped_data.pivot(index='income_group', columns='Gender', values='Price ($)')

# Display the pivot table
print(pivot_data)

# Heatmap to visualize the variation
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title('Average Selling Price by Gender and Income Group')
plt.xlabel('Gender')
plt.ylabel('Income Group')
plt.show()
```

```python
#7. What is the distribution of car prices by region, and how does the number of cars sold vary by region?

data = data.dropna(subset=['Dealer_Region', 'Price ($)'])

# 1. Distribution of car prices by region (Box Plot)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Dealer_Region', y='Price ($)', data=data, palette="coolwarm")
plt.title('Distribution of Car Prices by Region')
plt.xlabel('Region')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Number of cars sold by region (Bar Chart)
region_counts = data['Dealer_Region'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=region_counts.index, y=region_counts.values, palette="viridis")
plt.title('Number of Cars Sold by Region')
plt.xlabel('Region')
plt.ylabel('Number of Cars Sold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
#8. How does the average car price differ between cars with different engine sizes?

avg_price_by_engine = df.groupby('Engine')['Price ($)'].mean().reset_index()

    # Rename columns for clarity
avg_price_by_engine.columns = ['Engine Size (cc)', 'Average Selling Price']

    # Sort by engine size for better visualization
avg_price_by_engine = avg_price_by_engine.sort_values(by='Engine Size (cc)')

    # Display the results
print("\nAverage Selling Price by Engine Size:")
print(avg_price_by_engine)

    # Plot the results
plt.figure(figsize=(10, 6))
plt.plot(avg_price_by_engine['Engine Size (cc)'], avg_price_by_engine['Average Selling Price'], marker='o', color='skyblue')
plt.title('Average Car Price by Engine Size')
plt.xlabel('Engine Size (cc)')
plt.ylabel('Average Selling Price')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#9. How do car prices vary based on the customer’s annual income bracket.

bins = [0, 50000, 100000, 200000, 500000, float('inf')]
labels = ['Low (<50k)', 'Lower-Middle (50k-100k)', 'Middle (100k-200k)', 'Upper-Middle (200k-500k)', 'High (>500k)']
df['income_bracket'] = pd.cut(df['Annual Income'], bins=bins, labels=labels, right=False)

    # Group by income bracket and calculate average selling price
avg_price_by_income = df.groupby('income_bracket')['Price ($)'].mean().reset_index()

    # Rename columns for clarity
avg_price_by_income.columns = ['Income Bracket', 'Average Selling Price']

    # Display the results
print("\nAverage Selling Price by Income Bracket:")
print(avg_price_by_income)

    # Plot the results
plt.figure(figsize=(10, 6))
plt.bar(avg_price_by_income['Income Bracket'], avg_price_by_income['Average Selling Price'], color='skyblue', edgecolor='black')
plt.title('Car Prices by Customer Annual Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Average Selling Price')
plt.tight_layout()
plt.show()

```

```python
#_10  What are the top 5 car models with the highest number of sales, and how does their price distribution look.

top_models = data['Model'].value_counts().head(5).index

# Filter the data for the top models
top_models_data = data[data['Model'].isin(top_models)]

# Display the top 5 models and their sales counts
top_model_counts = data['Model'].value_counts().head(5)
print("Top 5 Car Models by Sales:")
print(top_model_counts)

# Plot a box plot for price distribution of the top models
plt.figure(figsize=(12, 6))
sns.boxplot(x='Model', y='Price ($)', data=top_models_data, order=top_models, palette="coolwarm")
plt.title('Price Distribution of Top 5 Car Models')
plt.xlabel('Car Model')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 11. How does car price vary with engine size across different car colors, and which colors have the highest price
#variation?

data = df.dropna(subset=['Color', 'Engine', 'Price ($)'])

# Ensure columns are numeric where applicable
data['engine_size'] = pd.to_numeric(df['Engine'], errors='coerce')
data['selling_price'] = pd.to_numeric(df['Price ($)'], errors='coerce')

# 1. Scatter plot to show how car price varies with engine size across different colors
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Engine', y='Price ($)', hue='Color', data=df, palette='tab20', alpha=0.7)
plt.title('Car Price vs Engine Size by Color')
plt.xlabel('Engine Size')
plt.ylabel('Selling Price')
plt.legend(title='Car Color', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# 2. Calculate price variation (standard deviation) for each color
price_variation = df.groupby('Color')['Price ($)'].std().reset_index()

# Sort the price variation by highest to lowest
price_variation_sorted = price_variation.sort_values(by='Price ($)', ascending=False)

# 3. Bar plot to visualize the price variation across colors
plt.figure(figsize=(12, 6))
sns.barplot(x='Color', y='Price ($)', data=price_variation_sorted, palette="viridis")
plt.title('Price Variation by Car Color')
plt.xlabel('Car Color')
plt.ylabel('Price Variation (Standard Deviation)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print the color with the highest price variation
print(price_variation_sorted.head())
```

```python
data = df.dropna(subset=['Price ($)', 'Engine', 'Annual Income'])

# Ensure relevant columns are numeric
data['selling_price'] = pd.to_numeric(df['Price ($)'], errors='coerce')
data['engine_size'] = pd.to_numeric(df['Engine'], errors='coerce')
data['annual_income'] = pd.to_numeric(df['Annual Income'], errors='coerce')


```

```python
# Calculate the correlation matrix for car price, engine size, and annual income
correlation_matrix = df[['Price ($)', 'Annual Income']].corr()

# Display the correlation matrix
print(correlation_matrix)

```

```python
# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Between Car Price, Annual Income')
plt.show()

# Pair plot to visualize the relationships between the features
sns.pairplot(df[['Price ($)', 'Engine', 'Annual Income']], kind='scatter')
plt.suptitle('Pairwise Relationships Between Car Price, Engine Size, and Annual Income', y=1.02)
plt.show()
```
